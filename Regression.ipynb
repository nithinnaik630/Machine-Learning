```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns

# Data
x = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)
y = np.array([1.5, 1.9, 3.0, 3.6, 4.2, 5.0])

model = LinearRegression()
model.fit(x, y)

intercept = model.intercept_
slope = model.coef_[0]

x_new = np.array([[7]])
prediction = model.predict(x_new)[0]

#visulation map
plt.scatter(x, y, color="blue", label="Data points")          # original data
plt.plot(x, model.predict(x), color="red", label="Regression line")  # regression line
plt.scatter(7, prediction, color="green", label="Prediction at X=7") # prediction point
plt.xlabel("Advertising Spend (₹000s)")
plt.ylabel("Sales (₹000s)")
plt.legend()
plt.show()


print(f" linear equation = {intercept:.2f} + {slope:.2f}")
print(f" precdication ads on speed 7000 for sale for next: {prediction:.2f}")
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# data

x = np.array([
    [1000, 1, 10],
    [1300, 2, 8],  #sq_ft, bedrooms, years old
    [1600, 3, 5]])
y = np.array([8, 20, 50])

# fit regression

model = LinearRegression()
model.fit(x, y)

#paremeter

intercept = model.intercept_
slope = model.coef_

#prediction

y_new = model.predict(x)

# prediction of new house price

x_new = np.array([[2000, 3, 10]])
prediction = model.predict(x_new)

print("predicted on new house price in lakhs :", prediction)
print(mean_squared_error(y,y_new))
print(r2_score(y, y_new))

```
