```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules


# load dataset

data = pd.read_excel("Online Retail.xlsx")

# data clearning

data = data[data['InvoiceNo'].astype(str).str.startswith('C')==False]
data = data[data['Quantity']>0]

# create basket

basket = (data[data['Country'] == 'United Kingdom']
          .groupby(['InvoiceNo','Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))

#convert binary

basket = (basket > 0).astype(int)

# apply aprior

frequent_items = apriori(basket , min_support = 0.02, use_colnames = True)

# apply rules

rules = association_rules(frequent_items,metric = 'confidence', min_threshold  = 0.2)

# sort the rules 

rules = rules.sort_values(['lift','confidence'], ascending = False)

print(rules[['support','confidence','lift']].head(10))

```
