```python
# University – Course Enrollment Pattern Analysis
```

```python
import pandas as pd
from mlxtend.frequent_patterns import fpgrowth, association_rules
from mlxtend.preprocessing import TransactionEncoder

# create a dataframe

data = {
    "Student_ID": ["S101", "S102", "S103", "S104", "S105"],
    "Courses_Enrolled": [
        ["Math", "Physics", "Chemistry"],
        ["Math", "Computer Science"],
        ["Biology", "Chemistry"],
        ["Math", "Physics"],
        ["Computer Science", "Math", "Statistics"]
    ]
}

df = pd.DataFrame(data)

# apply transactionencoder

te  = TransactionEncoder()
te_encoder = te.fit(df["Courses_Enrolled"]).transform(df["Courses_Enrolled"])
te_da = pd.DataFrame(te_encoder, columns = te.columns_)

# apply fp growth

growth = fpgrowth(te_da, min_support = 0.01, use_colnames=True)

print(growth)

# apply rules

rules = association_rules(growth, metric = 'lift', min_threshold= 1.0)
print(rules[["confidence", "lift", "support"]])

co_occurance = growth.explode("itemsets")["itemsets"].value_counts()
print(co_occurance)

interdisciplinary = rules[(rules['lift'] > 1.2) & (rules['confidence'] > 0.6)]

print("\nPotential Interdisciplinary Course Bundles:\n", interdisciplinary[[ "support", "confidence", "lift"]])
```

```python
# Medical Prescription Pattern Analysis
```

```python
import pandas as pd
from mlxtend.frequent_patterns import fpgrowth, association_rules
from mlxtend.preprocessing import TransactionEncoder

# create a datasete 

data = {
    'PrescriptionID': [1, 2, 3, 4, 5],
    'Drugs': [
        ['Paracetamol', 'Ibuprofen'],
        ['Amoxicillin', 'Paracetamol'],
        ['Ibuprofen', 'Cetirizine'],
        ['Paracetamol', 'Ibuprofen', 'Cetirizine'],
        ['Amoxicillin', 'Ibuprofen']
    ]
}

df = pd.DataFrame(data)


# applying transaction encoder

te = TransactionEncoder()
encoder = te.fit(df['Drugs']).transform(df['Drugs'])
te_enc = pd.DataFrame(encoder , columns =te.columns_)
print(te_enc)

# applying fp growth

frequent_item = fpgrowth(te_enc , min_support = 0.03,use_colnames = True)
print("frequent_item/n:",frequent_item )

# apply association rules 
rules = association_rules( frequent_item , metric = "lift" , min_threshold = 1)
print(rules[["confidence", "support", "lift"]])

co_occurance = frequent_item.explode("itemsets")["itemsets"].value_counts()
print(co_occurance)

rare_combos = frequent_item[frequent_item['support'] < 0.2]

print("\nUnusual Prescription Patterns (Rare Combos):")
print(rare_combos)
```
