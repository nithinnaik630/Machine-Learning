```python
# Spam detection
```

```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Load dataset file

data = pd.read_csv("SMSSpamCollection", sep = "\t", names = ["label","message"])

x = data['message']
y = data['label'] = data['label'].map({'ham':0, 'spam':1})

# train_test_split data

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

# vectorzition

vectorzier = TfidfVectorizer( stop_words = 'english', max_features = 6000)
x_train_vec = vectorzier.fit_transform(x_train)
x_test_vec =  vectorzier.transform(x_test)

# model

model = MultinomialNB(alpha = 1.0)
model.fit(x_train_vec, y_train)

pred = model.predict(x_test_vec)

# elevation

print("spam->1, Not-spam->0:", pred)
print("Accuracy: {:.2f}%". format(accuracy_score(y_test, pred)*100))
print("confusion_matrix:\n", confusion_matrix(y_test, pred))
print("classifiaction_report:\n",classification_report(y_test, pred))
```

```python
# News Article Topic Classification using Naive Bayes
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# load dataset file 
data = pd.read_csv("bbc-text.csv")

print(data.shape)


# vectorzier

vec = TfidfVectorizer( stop_words = 'english', max_features =9000)
x = vec.fit_transform(data['text'])
y = data['category'].values

# train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)

# model 

model = MultinomialNB(alpha = 1.0)
model.fit(x_train, y_train)
pred = model.predict(x_test)

# elevation

print("prediction:", pred)
print("Accuracy: {:.2f}%". format(accuracy_score(y_test, pred)*100))
print("confusion_matrix:\n", confusion_matrix(y_test, pred))
print("classifiaction_report:\n",classification_report(y_test, pred))

# new prediction

sample_text = ["The company announced a major breakthrough in AI technology."]
sample_features = vec.transform(sample_text)
print("Predicted Category:", model.predict(sample_features)[0])
```
