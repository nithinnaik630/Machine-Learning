```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns

# Data
x = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)
y = np.array([1.5, 1.9, 3.0, 3.6, 4.2, 5.0])

model = LinearRegression()
model.fit(x, y)

intercept = model.intercept_
slope = model.coef_[0]

x_new = np.array([[7]])
prediction = model.predict(x_new)[0]

#visulation map
plt.scatter(x, y, color="blue", label="Data points")          # original data
plt.plot(x, model.predict(x), color="red", label="Regression line")  # regression line
plt.scatter(7, prediction, color="green", label="Prediction at X=7") # prediction point
plt.xlabel("Advertising Spend (₹000s)")
plt.ylabel("Sales (₹000s)")
plt.legend()
plt.show()


print(f" linear equation = {intercept:.2f} + {slope:.2f}")
print(f" precdication ads on speed 7000 for sale for next: {prediction:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.pipeline import Pipeline


df = pd.read_csv('nonlinear_regression.csv')
print(df.head())

#data
X = df[['x']].values
Y = df[['y']].values

#split into test/train
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = 42)
                                                   
#linear regression
model = LinearRegression()
model.fit(x_train, y_train)
X_model = model.predict(x_test)
print(r2_score(y_test, X_model))

#polynomialregression
poly = Pipeline([
    ("poly_features", PolynomialFeatures(degree=2, include_bias=False)),
    ("lin_reg", LinearRegression())
])
poly.fit(x_train, y_train)
x_poly = poly.predict(x_test)
print(r2_score(y_test, x_poly))


#vasulization
x_plot = np.linspace(X.min(), X.max(), 200).reshape(-1, 1)
y_lin_plot = model.predict(x_plot)
y_poly_plot = poly.predict(x_plot)

plt.scatter(X, Y, color="red", label="data")
plt.plot(x_plot, y_lin_plot, color="green", label="linear_fit")
plt.plot(x_plot, y_poly_plot, color= "pink", label="polynomial")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures


# data 
x = np.array([2005, 2008, 2010, 2013, 2016, 2019, 2022]).reshape(-1, 1) # years of old house
y = np.array([15, 25, 20, 35, 50, 65, 80]) #price of house in lakhs

# tansofrom features into polynimal degree
model = PolynomialFeatures(degree= 3)
X_model = model.fit_transform(x)
y_model = LinearRegression()
y_model.fit(X_model, y)
pre = y_model.predict(X_model)


# linear regression
model_1 = LinearRegression()
model_1.fit(x, y)
pre_1 = model_1.predict(x)

# future prediction

future_year = np.array([2026, 2030, 2033]).reshape(-1, 1)
poly = model.transform(future_year)
future_pred = y_model.predict(poly)


for year, price in zip(future_year.flatten(), future_pred):
    print(f"Predicted price in {year}: {price:.2f} lakhs")

#vasulization
plt.scatter(x, y, color="blue", label="data")
plt.plot(x, pre, color = "red", label="polynomial")
plt.scatter(future_year, future_pred, color="purple", marker="x", s=100, label="Future Predictions")
plt.xlabel("x")
plt.ylabel("y")
plt.title("House price, current price and future price preduction")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import KFold, cross_val_score

# Data
x = np.array([5, 10, 15, 20, 25, 30, 35]).reshape(-1, 1)  # days
y = np.array([2.0, 4.5, 9.0, 15.0, 22.0, 28.0, 31.0])     # height in cm

# Polynomial Regression (degree 4)
degree = 4
poly = PolynomialFeatures(degree=degree)
X_poly = poly.fit_transform(x)
poly_model = LinearRegression()

# K-Fold Cross Validation (k=5)
kf = KFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_val_score(poly_model, X_poly, y, cv=kf, scoring="r2")

print("R² scores for each fold:", scores)
print("Average R² score:", scores.mean())

# Fit on full dataset for visualization
poly_model.fit(X_poly, y)
y_poly_pred = poly_model.predict(X_poly)

# Linear Regression for comparison
lin_model = LinearRegression()
lin_model.fit(x, y)
y_lin_pred = lin_model.predict(x)

# Visualization
plt.scatter(x, y, color="blue", label="Data")
plt.plot(x, y_poly_pred, color="red", label=f"Polynomial Regression (deg={degree})")
plt.plot(x, y_lin_pred, color="green", label="Linear Regression")
plt.xlabel("Days")
plt.ylabel("Height (cm)")
plt.title("Plant Growth with Polynomial Regression and K-Fold Validation")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# 1. Generate synthetic data
np.random.seed(42)
X = np.linspace(-5, 5, 100).reshape(-1, 1)
y_true = 3*X**3 - 2*X**2 + X
noise = np.random.normal(0, 10, size=y_true.shape)
y = y_true + noise

# 2. Fit polynomial models of degree 1, 2, 3, 5
degrees = [1, 2, 3, 5]
models = {}
rmse_scores = {}

for d in degrees:
    poly = PolynomialFeatures(degree=d, include_bias=False)
    X_poly = poly.fit_transform(X)
    model = LinearRegression().fit(X_poly, y)
    y_pred = model.predict(X_poly)
    rmse = np.sqrt(mean_squared_error(y, y_pred))
    models[d] = (model, y_pred)
    rmse_scores[d] = rmse

# 3. Plot all models
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color="black", label="Data (noisy)")
colors = {1:"red", 2:"green", 3:"blue", 5:"purple"}
for d in degrees:
    plt.plot(X, models[d][1], color=colors[d], label=f"Degree {d} (RMSE={rmse_scores[d]:.2f})")
plt.legend()
plt.title("Polynomial Regression Fits")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

# 4. Print RMSE results
print("RMSE by degree:")
for d in degrees:
    print(f"Degree {d}: RMSE = {rmse_scores[d]:.2f}")

# 5. Identify underfitting/overfitting
print("\nDiagnosis:")
print("Degree 1 and 2 → Underfitting (cannot capture cubic trend).")
print("Degree 3 → Best fit (matches true generating function).")
print("Degree 5 → Slightly lower RMSE but adds unnecessary complexity → Overfitting risk.")
```
